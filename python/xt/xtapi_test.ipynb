```python
import numpy as np
price=np.random.rand(1,10).tolist()[0]
new_price=[round(x*10000,2) for x in price]
data=[{'price':i,'amount':0.01,'type':1} for i in new_price]
kws=get_auth_payload({'market':'btc_usdt',
                         'data':data,
                         })
status,orders,_=sra.palce_orders(kwargs=kws)
```
